## Problem 3: Preprocessing Pipeline

In this problem, I implemented a manual preprocessing function for MNIST without using any built-in `preprocess_input` helper. The function performs three required steps: pixel normalization, reshaping to add the channel dimension, and one-hot encoding of the labels.

To verify correctness, I printed the shapes, data types, value ranges, and sample values before and after preprocessing. This confirms that the images are converted from `uint8` to `float32`, scaled from the range `0-255` to `0.0-1.0`, reshaped from `(N, 28, 28)` to `(N, 28, 28, 1)`, and that the labels are converted into one-hot vectors of length `10`.

````python
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

def preprocess_mnist(images, labels):
    images = images.astype(np.float32) / 255.0
    images = images.reshape(images.shape[0], images.shape[1], images.shape[2], 1)
    labels = tf.keras.utils.to_categorical(labels, num_classes=10).astype(np.float32)
    return images, labels

sample_index = 0

print(f"Before normalization dtype: {mnist_x_train.dtype}")
print(f"Before normalization shape: {mnist_x_train.shape}")
print(f"Before normalization min/max: {mnist_x_train.min()} {mnist_x_train.max()}")
print(f"Before normalization sample pixels: {mnist_x_train[sample_index, 0, :10].tolist()}")
print(f"Before one-hot sample label: {int(mnist_y_train[sample_index])}")

mnist_x_processed, mnist_y_processed = preprocess_mnist(mnist_x_train, mnist_y_train)

print(f"After normalization dtype: {mnist_x_processed.dtype}")
print(f"After preprocessing image shape: {mnist_x_processed.shape}")
print(f"After normalization min/max: {mnist_x_processed.min()} {mnist_x_processed.max()}")
print(f"After normalization sample pixels: {mnist_x_processed[sample_index, 0, :10, 0].tolist()}")
print(f"After one-hot label shape: {mnist_y_processed.shape}")
print(f"After one-hot sample label: {mnist_y_processed[sample_index].tolist()}")
````

## Problem 4: Data Augmentation Pipeline

For this problem, I built a training-time augmentation pipeline for CIFAR-10 using three random transformations: horizontal flip with probability `0.5`, random rotation in the range of approximately `+/-10` degrees, and random zoom up to `10%`. These augmentations are appropriate for CIFAR-10 because they create useful variation while generally preserving the semantic class of the object.

I then applied the augmentation pipeline to 5 CIFAR-10 training images and displayed each original image alongside 3 different augmented versions in a `5 x 4` grid. The resulting figure was saved as `augmentation_demo.png`.

````python
import tarfile
import pickle

os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

archive_path = os.path.join(os.getcwd(), ".keras", "datasets", "cifar-10-batches-py-target_archive")

batch_arrays = []
label_arrays = []

with tarfile.open(archive_path, "r:gz") as tar:
    for batch_id in range(1, 6):
        member = tar.getmember(f"cifar-10-batches-py/data_batch_{batch_id}")
        with tar.extractfile(member) as file_obj:
            batch = pickle.load(file_obj, encoding="bytes")
        batch_arrays.append(batch[b"data"])
        label_arrays.extend(batch[b"labels"])

cifar_x_train_manual = np.vstack(batch_arrays).reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1).astype(np.float32) / 255.0

augmentation_pipeline = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal", seed=42),
    tf.keras.layers.RandomRotation(10 / 360.0, fill_mode="nearest", seed=42),
    tf.keras.layers.RandomZoom(height_factor=(-0.1, 0.1), width_factor=(-0.1, 0.1), fill_mode="nearest", seed=42)
], name="cifar10_augmentation")

sample_indices = [0, 1, 2, 3, 4]
sample_images = cifar_x_train_manual[sample_indices]

fig, axes = plt.subplots(5, 4, figsize=(10, 12))
fig.suptitle("CIFAR-10 Data Augmentation Demo", fontsize=16)
column_titles = ["Original", "Augmented 1", "Augmented 2", "Augmented 3"]

for row, image in enumerate(sample_images):
    variants = [image]
    image_tensor = tf.expand_dims(image, axis=0)
    for _ in range(3):
        augmented_image = augmentation_pipeline(image_tensor, training=True)[0].numpy()
        variants.append(np.clip(augmented_image, 0.0, 1.0))
    for col, variant in enumerate(variants):
        axes[row, col].imshow(variant)
        if row == 0:
            axes[row, col].set_title(column_titles[col], fontsize=10)
        axes[row, col].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig("augmentation_demo.png", dpi=200, bbox_inches="tight")
plt.show()

print("Saved figure: augmentation_demo.png")
print("Figure layout: 5 rows x 4 columns")
print(f"Loaded CIFAR-10 samples: {cifar_x_train_manual.shape[0]}")
print("Augmentation pipeline: horizontal flip (p=0.5), random rotation (+/-10 degrees), random zoom up to 10%")
````

## Written Justification

Data augmentation should be applied only to the training set because its purpose is to help the model learn to generalize from more varied examples. By showing slightly modified versions of the training images, we reduce overfitting and make the model more robust to small changes in object appearance, position, and orientation.

Augmentation should never be applied to the validation or test set because those sets are meant to represent the original unseen data distribution. If we augment validation or test samples, then the evaluation would no longer measure performance on the true dataset. This would make the reported accuracy less reliable and could lead to unfair or inconsistent comparisons between models.

## Q1

### What does the channel dimension represent in a tensor of shape `(N, H, W, C)`? Explain with reference to both a grayscale image and an RGB image.

In a tensor of shape `(N, H, W, C)`, the channel dimension `C` represents how many values are stored for each pixel location. The dimensions `H` and `W` describe the height and width of the image, while `C` describes the depth of information at each pixel. The dimension `N` represents the number of images in the batch.

For a grayscale image, each pixel has only one intensity value, so the channel dimension is `1`. That is why a batch of grayscale MNIST images is written as `(N, 28, 28, 1)`. For an RGB image, each pixel has three separate values corresponding to red, green, and blue, so the channel dimension is `3`. That is why a batch of CIFAR-10 images is written as `(N, 32, 32, 3)`.

## Q2

### CIFAR-10 images are `32 x 32` pixels. If you were training a CNN on `1024 x 1024` satellite images, what data loading strategies would you use to avoid running out of memory? Name at least two techniques and explain how each one helps.

If I were training a CNN on `1024 x 1024` satellite images, one important strategy would be to use a smaller batch size. Large images consume much more GPU and RAM memory than CIFAR-10 images, so reducing the batch size helps fit the data into memory during training.

A second useful strategy would be patch-based loading or tiling. Instead of loading the full high-resolution image at once, the image can be split into smaller patches such as `256 x 256` or `512 x 512`. This reduces memory usage and also allows the model to learn from local regions without storing the full image tensor in memory.

Another helpful technique is on-the-fly data loading using generators or `tf.data` pipelines. This loads images from disk in small batches only when needed, instead of keeping the full dataset in RAM. Prefetching and efficient streaming can also improve throughput without exhausting memory.

## Q3

### Suppose a student applies normalisation to the test set using the mean and standard deviation computed from the test set itself. What is wrong with this approach?

Normalizing the test set using the mean and standard deviation computed from the test set itself is wrong because it introduces data leakage. The test set is supposed to remain completely unseen during model development, and its statistics should not be used to influence preprocessing decisions.

The correct approach is to compute the normalization parameters only from the training set and then apply those same values to the validation and test sets. This keeps evaluation fair and ensures that model performance reflects how the system would behave on truly unseen data.

## Problem 1

### Manual 2D Convolution (NumPy only)

In this problem, I implemented a `conv2d(image, kernel, stride, padding)` function using only NumPy. The function accepts a 2D grayscale image and a 2D kernel, supports integer stride and zero-padding, and returns the correct 2D feature map by sliding the kernel over the image with nested loops.

I tested the implementation on the provided `5 x 5` image using the Sobel-X kernel with `stride = 1` and `padding = 0`. I also verified the output shape manually using the convolution output-size formula from the primer.

````python
test_image = np.array([
    [3, 1, 0, 2, 4],
    [1, 5, 3, 2, 1],
    [0, 2, 6, 4, 3],
    [2, 3, 1, 5, 2],
    [1, 0, 2, 3, 4]
], dtype=np.float32)

sobel_x_kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

def conv2d(image, kernel, stride=1, padding=0):
    if padding > 0:
        padded_image = np.pad(image, ((padding, padding), (padding, padding)), mode="constant")
    else:
        padded_image = image.copy()

    image_height, image_width = padded_image.shape
    kernel_height, kernel_width = kernel.shape

    output_height = ((image_height - kernel_height) // stride) + 1
    output_width = ((image_width - kernel_width) // stride) + 1
    feature_map = np.zeros((output_height, output_width), dtype=np.float32)

    for i in range(output_height):
        for j in range(output_width):
            row_start = i * stride
            row_end = row_start + kernel_height
            col_start = j * stride
            col_end = col_start + kernel_width
            patch = padded_image[row_start:row_end, col_start:col_end]
            feature_map[i, j] = np.sum(patch * kernel)

    return feature_map

feature_map_output = conv2d(test_image, sobel_x_kernel, stride=1, padding=0)

manual_output_height = ((5 - 3 + 2 * 0) // 1) + 1
manual_output_width = ((5 - 3 + 2 * 0) // 1) + 1

print("Feature map:")
print(feature_map_output)
print(f"Output shape: {feature_map_output.shape}")
print(f"Manual output shape: {(manual_output_height, manual_output_width)}")
````

## Numerical Output Of `conv2d`

The computed feature map is:

```text
[[ 7. -3. -3.]
 [13.  3. -7.]
 [ 5.  9.  1.]]
```

The output shape is `(3, 3)`, which matches the manually calculated shape using the formula:

```text
Output = floor((Input - Kernel + 2 * Padding) / Stride) + 1
       = floor((5 - 3 + 0) / 1) + 1
       = 3
```


## Observation

The NumPy-based convolution function produced the expected `3 x 3` feature map for the given `5 x 5` input image and `3 x 3` Sobel-X kernel. Since stride was `1` and padding was `0`, the kernel could move across only the valid spatial positions, which reduced the output size from `5 x 5` to `3 x 3`.

The positive and negative values in the output indicate that the Sobel-X kernel is responding to horizontal intensity changes across the image, which is consistent with its role as an edge detector in the x-direction.

## Problem 2

### Output Size Derivation

For all parts below, I used the convolution output-size formula from the primer:

```text
Output = floor((Input - Kernel + 2 x Padding) / Stride) + 1
```

The same formula is applied independently to both height and width because all given inputs are square.

## Calculations

### (a) Input: `28 x 28`, Kernel: `5 x 5`, Padding: `0`, Stride: `1`

```text
Output = floor((28 - 5 + 2 x 0) / 1) + 1
       = floor(23 / 1) + 1
       = 23 + 1
       = 24
```

So, the output spatial size is **`24 x 24`**.

### (b) Input: `28 x 28`, Kernel: `3 x 3`, Padding: `1`, Stride: `1`

```text
Output = floor((28 - 3 + 2 x 1) / 1) + 1
       = floor((28 - 3 + 2) / 1) + 1
       = floor(27 / 1) + 1
       = 27 + 1
       = 28
```

So, the output spatial size is **`28 x 28`**.

### (c) Input: `32 x 32`, Kernel: `3 x 3`, Padding: `0`, Stride: `2`

```text
Output = floor((32 - 3 + 2 x 0) / 2) + 1
       = floor(29 / 2) + 1
       = floor(14.5) + 1
       = 14 + 1
       = 15
```

So, the output spatial size is **`15 x 15`**.

### (d) Two consecutive Conv2D layers

First layer: `K = 3`, `P = 1`, `S = 1`, input = `32 x 32`

```text
Output_1 = floor((32 - 3 + 2 x 1) / 1) + 1
         = floor((32 - 3 + 2) / 1) + 1
         = floor(31 / 1) + 1
         = 31 + 1
         = 32
```

So after the first convolution, the output is **`32 x 32`**.

Second layer: `K = 3`, `P = 0`, `S = 1`, input = `32 x 32`

```text
Output_2 = floor((32 - 3 + 2 x 0) / 1) + 1
         = floor(29 / 1) + 1
         = 29 + 1
         = 30
```

So, the final output spatial size is **`30 x 30`**.

## Observation

These examples show how padding and stride directly control the output size of a convolution layer. Using padding `1` with a `3 x 3` kernel and stride `1` preserves the spatial size, while valid convolution with no padding reduces it. A stride greater than `1` shrinks the output more aggressively because the kernel moves in larger steps.

For stacked convolution layers, the output of one layer becomes the input of the next, so the size must be tracked step by step. This is important when designing CNN architectures because repeated valid convolutions can reduce feature-map size quickly.

## Problem 3

### Implement LeNet-5

I implemented the architecture exactly in the order specified in the question using TensorFlow/Keras:

```text
Conv2D(16 filters, 5x5, valid) -> Tanh -> AvgPool(2x2, stride 2) ->
Conv2D(16 filters, 5x5, valid) -> Tanh -> AvgPool(2x2, stride 2) ->
Flatten -> Dense(120) -> Tanh -> Dense(84) -> Tanh -> Dense(10, softmax)
```

The model builds correctly for input shape `(28, 28, 1)`. However, the exact architecture written in the prompt produces a total of `48,686` parameters, not `44,426`. I kept the implementation faithful to the provided layer description and documented the actual count from the model summary.

````python
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

lenet5_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(28, 28, 1)),
    tf.keras.layers.Conv2D(16, 5, padding="valid", activation="tanh"),
    tf.keras.layers.AveragePooling2D(pool_size=2, strides=2),
    tf.keras.layers.Conv2D(16, 5, padding="valid", activation="tanh"),
    tf.keras.layers.AveragePooling2D(pool_size=2, strides=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(120, activation="tanh"),
    tf.keras.layers.Dense(84, activation="tanh"),
    tf.keras.layers.Dense(10, activation="softmax")
], name="LeNet5")

lenet5_summary_lines = []
lenet5_model.summary(print_fn=lenet5_summary_lines.append)
print("\n".join(lenet5_summary_lines))
print(f"Total params: {lenet5_model.count_params()}")

first_conv_params = (5 * 5 * 1 + 1) * 16
print(f"First Conv2D params: {first_conv_params}")
````

## Parameter Count Workings + Explanation

### First Conv2D layer parameter count

For the first convolution layer, the formula is:

```text
Params = (K_h x K_w x C_in + 1) x C_out
       = (5 x 5 x 1 + 1) x 16
       = (25 + 1) x 16
       = 26 x 16
       = 416
```

So, the first Conv2D layer has **`416`** trainable parameters.

### Why AvgPooling was used in LeNet-5 but MaxPooling is more common today

Average pooling was used in the original LeNet-5 because it provides a smooth form of downsampling by taking the mean value in each local region. In early CNN designs, this was a natural choice because it reduced spatial resolution while preserving an overall summary of the activation map.

MaxPooling is more common in modern CNNs because it tends to preserve the strongest or most important local activation, which often makes feature extraction more robust. In practice, MaxPooling usually helps later layers focus on the most discriminative patterns, while AveragePooling can sometimes smooth out strong responses too much.

## Problem 4

### Design A Custom CNN For CIFAR-10

Below is the architecture sketch for my custom CNN. It uses three convolutional blocks, each following the pattern `Conv2D -> BatchNorm -> ReLU -> Pooling`, followed by a compact classification head with dropout. I used `GlobalAveragePooling2D` instead of flattening the full feature map because it reduces parameter growth and encourages the network to summarize spatial features more efficiently.

```text
Input (32 x 32 x 3)
  -> Conv2D(48, 3x3, same, no bias)
  -> BatchNorm
  -> ReLU
  -> MaxPool(2x2)
  -> Conv2D(96, 3x3, same, no bias)
  -> BatchNorm
  -> ReLU
  -> MaxPool(2x2)
  -> Conv2D(192, 3x3, same, no bias)
  -> BatchNorm
  -> ReLU
  -> MaxPool(2x2)
  -> GlobalAveragePooling2D
  -> Dense(256, ReLU)
  -> Dropout(0.4)
  -> Dense(10, Softmax)
```

I increased the number of filters gradually from `48` to `192` so that deeper layers can learn richer and more abstract representations while early layers remain relatively lightweight. Batch normalization is included after every convolution to stabilize training, and dropout is used in the dense head to reduce overfitting. The total parameter count stays within the required range, giving the model enough capacity for CIFAR-10 without becoming unnecessarily large.

````python
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

custom_cifar10_cnn = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(32, 32, 3)),
    tf.keras.layers.Conv2D(48, 3, padding="same", use_bias=False),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Conv2D(96, 3, padding="same", use_bias=False),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Conv2D(192, 3, padding="same", use_bias=False),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(10, activation="softmax")
], name="CustomCIFAR10CNN")

custom_summary_lines = []
custom_cifar10_cnn.summary(print_fn=custom_summary_lines.append)
print("\n".join(custom_summary_lines))
print(f"Total params: {custom_cifar10_cnn.count_params()}")
````

## Q1

### Compare the parameter efficiency of two stacked `3 x 3` Conv layers versus one `5 x 5` Conv layer on the same input with the same number of filters. Which uses fewer parameters? Show numerical proof and explain any other advantages of the smaller kernel approach.

Two stacked `3 x 3` convolution layers are more parameter-efficient than one `5 x 5` convolution layer when they use the same channel width and number of filters. A single `5 x 5` convolution has parameter count `(5 x 5 x C + 1) x C = (25C + 1)C`, while two stacked `3 x 3` layers have `2 x ((3 x 3 x C + 1) x C) = 2 x (9C + 1)C`. Since `18C^2 + 2C` is smaller than `25C^2 + C` for practical channel sizes, the stacked `3 x 3` option uses fewer parameters.

For example, if `C = 64`, then one `5 x 5` layer has `(25 x 64 + 1) x 64 = 102,464` parameters. Two stacked `3 x 3` layers have `2 x ((9 x 64 + 1) x 64) = 73,856` parameters. Therefore, the two `3 x 3` layers use fewer parameters while covering a similar effective receptive field.

Another advantage of stacked smaller kernels is that they introduce an extra activation function between the layers, which increases the expressive power of the network. They also usually reduce computation and often lead to more efficient feature extraction in deep CNNs.

## Q2

### What is the role of Batch Normalisation in a CNN? Where in the layer stack should it be placed (before or after activation), and why? Mention at least two empirical benefits it provides during training.

Batch Normalisation helps stabilize intermediate activations during training by normalizing them across the mini-batch and then learning a trainable scale and shift. In CNNs, it is commonly placed after the convolution layer and before the activation function, so the activation receives normalized inputs. This ordering usually makes optimization more stable and keeps feature distributions better controlled.

One empirical benefit is faster convergence, because the network can often train with larger learning rates and smoother gradient flow. Another benefit is improved stability, which reduces sensitivity to poor initialization and helps avoid exploding or vanishing activations. It can also have a mild regularization effect that sometimes improves generalization.

## Q3

### Your custom CNN has a GlobalAveragePooling layer before the Dense head. What does this layer do geometrically? What would happen to the parameter count and spatial information if you replaced it with `Flatten`?

Geometrically, `GlobalAveragePooling` takes each feature map and averages all of its spatial values into one number. If the incoming tensor has shape `(H, W, C)`, then the output becomes `(C)`, because every channel is summarized by a single average activation. This keeps channel-level semantic information while collapsing the spatial dimensions.

If `Flatten` were used instead, then all spatial locations would be preserved as separate inputs to the Dense layer. That would increase the parameter count significantly because the classifier would connect to `H x W x C` inputs rather than just `C` inputs. Flatten retains more raw spatial detail, but it also makes the dense head much larger and increases the risk of overfitting.
